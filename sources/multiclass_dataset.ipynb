```python
import pandas as pd
```

```python
from datasets import load_dataset
```

```python
spiced_ds = load_dataset("copenlu/spiced")
```

```python
spiced_train_ds = spiced_ds["train"]
spiced_val_ds = spiced_ds["validation"]
spiced_test_ds = spiced_ds["test"]
```

```python
causality_train_labels = pd.read_csv("../data/aggregated-labels/causality/train-test-splits/causality_train.tsv", sep='\t')
causality_test_labels = pd.read_csv("../data/aggregated-labels/causality/train-test-splits/causality_test.tsv", sep='\t')
certainty_train_labels = pd.read_csv("../data/aggregated-labels/certainty/train-test-splits/certainty_train.tsv", sep='\t')
certainty_test_labels = pd.read_csv("../data/aggregated-labels/certainty/train-test-splits/certainty_test.tsv", sep='\t')
```

```python
train_ds = pd.DataFrame(columns=["finding", "causality", "certainty"])
for (_, labels) in causality_train_labels.iterrows():
    key = labels["id"]
    finding = spiced_train_ds.filter(lambda example: example['instance_id'] == key)
    if not finding:
        finding = spiced_val_ds.filter(lambda example: example['instance_id'] == key)
    if not finding:
        finding = spiced_test_ds.filter(lambda example: example['instance_id'] == key)
    certainty_labels = certainty_train_labels.loc[certainty_train_labels['id'] == key]
    print(certainty_labels)
    if finding is not None and certainty_labels is not None:
        train_ds.loc[len(train_ds)] = {
            "finding": finding["Paper Finding"][0],
            "causality": labels["label_paper_finding"],
            "certainty": certainty_train_labels.loc[certainty_train_labels['id'] == key, 'label_paper_finding'].iloc[0]
        }
        train_ds.loc[len(train_ds)] = {
            "finding": finding["News Finding"][0],
            "causality": labels["label_reported_finding"],
            "certainty":certainty_train_labels.loc[certainty_train_labels['id'] == key, 'label_reported_finding'].iloc[0]
        }
train_ds.to_csv('../temp/causality_cert_train.csv', index=True)
```

```python
test_ds = pd.DataFrame(columns=["finding", "causality", "certainty"])
for (_, labels) in causality_test_labels.iterrows():
    key = labels["id"]
    finding = spiced_train_ds.filter(lambda example: example['instance_id'] == key)
    if not finding:
        finding = spiced_val_ds.filter(lambda example: example['instance_id'] == key)
    if not finding:
        finding = spiced_test_ds.filter(lambda example: example['instance_id'] == key)
    certainty_labels = certainty_test_labels.loc[certainty_test_labels['id'] == key]
    print(certainty_labels)
    if finding is not None and certainty_labels is not None:
        test_ds.loc[len(test_ds)] = {
            "finding": finding["Paper Finding"][0],
            "causality": labels["label_paper_finding"],
            "certainty": certainty_test_labels.loc[certainty_test_labels['id'] == key, 'label_paper_finding'].iloc[0]
        }
        test_ds.loc[len(test_ds)] = {
            "finding": finding["News Finding"][0],
            "causality": labels["label_reported_finding"],
            "certainty":certainty_test_labels.loc[certainty_test_labels['id'] == key, 'label_reported_finding'].iloc[0]
        }
test_ds.to_csv('../temp/causality_cert_test.csv', index=True)
```
